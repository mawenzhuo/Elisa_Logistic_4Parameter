```python
import pandas as pd
import numpy as np
import os

def solve_for_x(y_value, A, B, C, D):
    """
    4PL 逆运算公式: x = C * ((A - y) / (y - D))**(1 / B)
    """
    try:
        # 将输入强制转为浮点数
        y = float(y_value)
        
        # 数学逻辑检查：(A-y)/(y-D) 必须大于 0 才能开 1/B 次方
        ratio = (A - y) / (y - D)
        
        if ratio <= 0:
            return np.nan
        
        # 计算 X
        x = C * (ratio**(1/B))
        return x
    except:
        # 捕获空值、文字、除以零等异常，统一返回 NaN
        return np.nan

def main():
    print("=== 4参数 Logistic (4PL) 矩阵逆运算工具 ===")
    
    # 1. 输入参数
    try:
        A = float(input("请输入参数 A (下渐近线): "))
        B = float(input("请输入参数 B (斜率): "))
        C = float(input("请输入参数 C (EC50): "))
        D = float(input("请输入参数 D (上渐近线): "))
    except ValueError:
        print("错误：输入的 A, B, C, D 必须是数字！")
        return

    # 2. 引导用户复制数据
    print("\n" + "="*40)
    print("请切换到 Excel，选中并【复制】(Ctrl+C) 您的 y 值数据区域。")
    print("数据可以包含空值，程序会自动处理。")
    input("确认数据已复制后，请在此处按【回车键】继续...")
    print("="*40)

    # 3. 从剪贴板读取数据
    try:
        # Excel 复制到剪贴板通常以制表符 \t 分隔
        df_y = pd.read_clipboard(sep='\t', header=None)
    except Exception as e:
        print(f"读取剪贴板失败: {e}")
        return

    if df_y.empty:
        print("错误：未在剪贴板中检测到有效数据！")
        return

    print(f"成功获取数据：{df_y.shape[0]} 行 x {df_y.shape[1]} 列。正在计算...")

    # 4. 执行逐单元格运算 (保持矩阵结构)
    # applymap 会对 DataFrame 中的每一个格子应用公式
    df_x = df_y.applymap(lambda val: solve_for_x(val, A, B, C, D))

    # 5. 输出为 Excel
    output_name = "calculate_X_values.xlsx"
    try:
        # 不保存行索引和表头，保持原始矩阵样式
        df_x.to_excel(output_name, index=False, header=False)
        print("\n" + "-"*30)
        print(f"计算完成！")
        print(f"输出文件：{os.path.abspath(output_name)}")
        print(f"提示：无效的 y 值、超量程值或空值已在结果中显示为 NAN。")
        
        # Windows 系统下尝试自动打开生成的文件
        if os.name == 'nt':
            os.startfile(output_name)
    except Exception as e:
        print(f"保存文件失败: {e}")

if __name__ == "__main__":
    main()
```
