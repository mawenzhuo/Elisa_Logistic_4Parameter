```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import io

def logistic4(x, A, B, C, D):
    """4参数logistic公式"""
    return D + (A - D) / (1 + (x / C)**B)

def main():
    print("=== 4PL 拟合工具 (针对加密 Excel) ===")
    print("步骤：")
    print("1. 打开你的加密 Excel。")
    print("2. 选中包含 'x' 和 'y' 表头及其数据的所有区域并【复制】。")
    print("3. 回到这里按下回车键。")
    input("\n确认已复制数据后，请按回车继续...")

    try:
        # 从剪贴板读取数据 (支持 Excel 的制表符分隔格式)
        df = pd.read_clipboard()
        
        # 强制将列名转为小写以防匹配失败
        df.columns = [str(col).strip().lower() for col in df.columns]
        
        if 'x' not in df.columns or 'y' not in df.columns:
            print("错误：复制的数据中未找到名为 'x' 和 'y' 的表头！")
            return

        # 去除空值并转换为浮点数
        df = df.dropna(subset=['x', 'y'])
        x_data = df['x'].astype(float).values
        y_data = df['y'].astype(float).values

        print(f"成功读取 {len(x_data)} 组数据。开始拟合...")

        # 设置初始猜测值 [A, B, C, D]
        # A: min(y), B: 1.0, C: median(x), D: max(y)
        p0 = [np.min(y_data), 1.0, np.median(x_data), np.max(y_data)]

        # 执行非线性拟合
        popt, pcov = curve_fit(logistic4, x_data, y_data, p0=p0, maxfev=10000)
        
        A, B, C, D = popt

        print("\n" + "="*40)
        print("拟合计算成功！结果如下：")
        print(f"A (下渐近线 Bottom): {A:.6f}")
        print(f"B (斜率 Hill Slope): {B:.6f}")
        print(f"C (拐点 EC50):       {C:.6f}")
        print(f"D (上渐近线 Top):    {D:.6f}")
        print("="*40)
        
    except Exception as e:
        print(f"\n运行出错：{e}")
        print("提示：请确保复制的内容包含 'x' 和 'y' 这两个单词作为表头。")

    input("\n处理完成，按回车键退出...")

if __name__ == "__main__":
    main()
```
